# Plot für 3D extreme eigenvalues

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def f(x):
    output =np.sin(x)/x**3-np.cos(x)/x**2
    return output
```

```python
def g(x):
    return 1/x**3 +1/x**2
```

```python
%matplotlib qt
fig=plt.figure()
n=1000
x_array = np.arange(n)/n*40
plt.plot(x_array, f(x_array), label='$f(x)$')
plt.plot(x_array, g(x_array), label=r'$1/x^3+1/x^2$')
plt.ylim(-0.1, 0.4)
plt.grid()
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.legend()
plt.show()
```
